```python
precios = ij.groupby(["id"]).sum()
precios["id"] = precios.index
precios.reset_index(inplace = True, drop = True)

precios.head()
```

```python
tabla_factura_final = fact.merge(precios, on = 'id', how = 'inner')
```

```python
tabla_factura_final["precio_final"] = tabla_factura_final['precio'] - tabla_factura_final["precio"]*tabla_factura_final['descuento']
```

```python
tabla_factura_final["precio_final"] = np.round(tabla_factura_final['precio_final'] + tabla_factura_final["precio_final"] * tabla_factura_final['impuesto']).astype(int)
```
